```python
import pytesseract
from PIL import Image
import pandas as pd

# Define a function to perform OCR on a specific area of the image
def ocr_section(image, box):
    """
    This function will handle OCR of a specific section of the image.
    :param image: PIL Image object
    :param box: Tuple (left, upper, right, lower) - the bounding box of the section
    :return: The string of text OCR recognized in the section
    """
    region = image.crop(box)
    text = pytesseract.image_to_string(region, lang='eng')
    return text.strip()

# Load the image from file
image = Image.open(image_path)

# Since OCR on the whole image at once is taking too long, we will try to OCR individual sections
# For example, we can define coordinates of the top section of the image to extract the store details
# These coordinates are just examples and will need to be adjusted to the specific sections of your image
# (left, upper, right, lower)
store_details_box = (50, 50, 700, 200)  # Adjust these values based on the actual image layout
store_details_text = ocr_section(image, store_details_box)

# Print out the OCR result of the section
store_details_text
```
